<a id='Set up virtual environment'><h2>Set up virtual environment</h2></a>
- <h3>Open terminal by searching "terminal" in spotlight.</h3>

<img src="https://farm8.staticflickr.com/7885/46173346444_e39d11a51a_b.jpg" width="524" height="339" alt="OpenTerminal">


- <h3>Let's check if conda is correctly installed. In terminal, type:</h3> 
    
    <code>conda -V</code>
    
<img src="https://farm8.staticflickr.com/7823/39933401383_2f65534ac9_b.jpg" width="524" height="73" alt="condaV">
 If conda is installed you should see the conda version. (This step should work if you correctly installed anaconda 3.7).
 
- <h3>Make the conda up to date by typing the following command to terminal:</h3>

    <code>conda update conda</code>
    
<img src="https://farm5.staticflickr.com/4915/46846226532_85977cb61c_b.jpg" width="549" alt="conda_update_conda">
    
- <h3>Create a virtual environment for our DataScienceNightSJSU</h3>

    <code>conda create -n DataScienceNightSJSU python=3.6.3 anaconda</code>
    
<img src="https://farm8.staticflickr.com/7825/39933401433_8729ab34bf_b.jpg" width="542" alt="condacreate1">
<img src="https://farm5.staticflickr.com/4870/46846226452_97c6d08c35_b.jpg" width="542" alt="condacreate2">

- <h3>Activate your virtual environment</h3>
    Activating a conda environment modifies the PATH and shell variables to point to the specific isolated Python set-up you created. 
    
    <code>source activate DataScienceNightSJSU</code>
    
    You can also see all available conda environments by typing <code>conda info -e </code>.

<img src="https://farm8.staticflickr.com/7876/46846226332_7766da8c99_b.jpg" width="532" alt="condinfoe">

<a id='Install python modules'>
<h1>Install python modules, e.g. Tensorflow, Keras, opencv</h1>
</a>
Anaconda is a convinent package manager that comes with most of the common python packages. 
However, to run my deep learning scripts, you need to separately install two deep learning packages (Keras and tensorflow) and a computer vision package, opencv, in your environment. 

- Note 1: It is very important to install the correct versions of these modules.  
- Note 2: You do NOT need GPU to follow this blog series as we will only use pre-trained network. 

To install the three packages, from your terminal run the following codes:

```
conda install tensorflow==1.9.0
conda install keras==2.1.2
conda install opencv==3.4.2 
```
